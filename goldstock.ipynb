```python
x_data = ["Date"]
y_data = ["High"]

plt.scatter(x_data, y_data, color='blue', marker='o', label='Scatter Plot')


plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example')

plt.legend()
plt.show()
```

```python
np.random.seed(4)
num_points = 1000
x_data = np.random.rand(num_points)
y_data = np.random.rand(num_points)
size_data = np.random.rand(num_points) * 50 
plt.scatter(x_data, y_data, s=size_data, alpha=0.2, c='blue', edgecolors='black')


plt.xlabel('Volume')
plt.ylabel('HIgh')
plt.title('Bubble Chart Example')

plt.show()
```

```python
x_data = np.arange(1, 6)
y1_data = np.array([1, 4, 6, 8, 9])
y2_data = np.array([0, 2, 5, 7, 8])


plt.fill_between(x_data, y1_data, y2_data, color='skyblue', alpha=0.4, label='Area Chart')


plt.xlabel('Volume')
plt.ylabel('High')
plt.title('Area Chart Example')

plt.legend()

plt.show()
```
